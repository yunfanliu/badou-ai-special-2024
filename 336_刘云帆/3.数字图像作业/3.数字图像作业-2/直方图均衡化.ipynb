1、导入模块

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

2、灰度直方图均衡化

```python
img=cv2.imread('lenna.png') #读取图片
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  #原图灰度化
dst_gray = cv2.equalizeHist(gray_img)   #灰度直方图均衡化
cv2.imshow('Gray histogram histogram equalization',dst_gray)
cv2.waitKey(0)
```

3、彩色直方图均衡化

```python
(b,g,r)=cv2.split(img)
bh = cv2.equalizeHist(b)
gh = cv2.equalizeHist(g)
rh = cv2.equalizeHist(r)
hist_img = cv2.merge((bh,gh,rh))
cv2.imshow('histogram histogram equalization',hist_img)
cv2.waitKey(0)
```

4、灰度直方图打印

```python
# 直方图
hist = cv2.calcHist([dst_gray],[0],None,[256],[0,256])

plt.figure()
plt.hist(dst_gray.ravel(), 256)
plt.show()
```
